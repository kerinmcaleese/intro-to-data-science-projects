```python
import pandas as pd
listings = pd.read_csv('Exercise Files/Airbnb_NYC_2019.csv')
listings.head()
```

```python
plt.hist(listings['price'])
plt.xlabel('price (in US dollars)')
plt.show()
```

```python
plt.hist(listings[ 'price' ], bins = np.arange(0, 1100, 40))
plt.xlabel ('price (in US dollars)') 
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings[ 'number_of_reviews' ])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.show()
```

```python
plt.scatter(x = listings [ 'price' ] , y = listings ['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0, 1100)
plt.show()
```

```python
plt.scatter(x = listings [ 'price' ] , y = listings ['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0, 1100)
plt.show()
```
