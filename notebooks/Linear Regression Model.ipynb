```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
market = pd.read_csv("Exercise Files/marketing_sales_data.csv")
market.head()
```

```python
market.shape
```

```python
market.isna().sum()
```

```python
sns.pairplot(market)
plt.show()
```

```python
train = market.sample(frac = 0.75, random_state=0)
test = market.drop(train.index)
```

```python
train.head()
```

```python
test.head()
```

```python
train_subset = train[["Radio", "Sales"]]
test_subset = test [["Radio", "Sales"]]
```

```python
train_subset.head()
```

```python
ols_market = ols(formula = "Sales ~ Radio", data = train_subset)
```

```python
model = ols_market.fit()
```

```python
model.summary()
```

```python
sns.regplot(x = "Radio", y = "Sales", data = train_subset)
plt.show
```

```python
residuals = model.resid
residuals
```

```python
sns. histplot(residuals)
plt.xlabel("Residual")
plt.show()
```

```python
sm.qqplot(residuals, line='s')
plt.show()
```
